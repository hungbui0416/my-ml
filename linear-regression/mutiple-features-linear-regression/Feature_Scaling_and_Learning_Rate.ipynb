### Feature scaling

#### z-score normalization

$$ x_j^{(i)} = \frac{x_j^{(i)} - \mu_j}{\sigma_j} $$

where

$$
\begin{align*}
\mu_j &= \frac{1}{m} \sum_{i=0}^{n-1}x_j^{(i)} \\
\sigma_j^2 &= \frac{1}{m} \sum_{i=0}^{n-1}(x_j^{(i)} - \mu_j) ^ 2
\end{align*}
$$

```python
def zscore_normalize_features(X):
    """ 
    Computes zscore normalized X
    Args:
        X (ndarray(n, d)): input variables
    Returns:
        X_norm (ndarray(n, d))  : zscored normalized X
        mu (ndarray(d,))        : mean of each feature
        sigma (ndarray(d,))     : standard deviation of each feature
    """
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

```python
X_normalized, mu, sigma = zscore_normalize_features(X_train)

fig, ax = plt.subplots(1, 2, figsize=(12, 3))

ax[0].scatter(X_train[:, 0], X_train[:, 3])
ax[0].set_xlabel(X_features[0]); ax[0].set_ylabel(X_features[3])
ax[0].set_title('unnormalized')
ax[0].axis('equal')

ax[1].scatter(X_normalized[:, 0], X_normalized[:, 3])
ax[1].set_xlabel(X_features[0])
ax[1].set_title('z-score normalized')
ax[1].axis('equal')

fig.suptitle('distribution of features before and after normalization')
plt.show()
```

```python
print(f"X_mu = {mu}\nX_sigma = {sigma}")
print(f"Peak to Peak range by column in Raw         {np.ptp(X_train, axis=0)}")
print(f"Peak to Peak range by column in Normalized  {np.ptp(X_normalized, axis=0)}")
```

#### $ \alpha = 1e^{-1} $

```python
alpha = 1e-1
iterations = 1000

w_norm, b_norm, J_hist_norm = gradient_descent_houses(X_normalized, y_train, initial_w, initial_b, alpha, iterations)
print(f"w,b found by gradient descent: w: {w_out}, b: {b_out:0.2f}")
```

```python
# prediction using normalized features
n = X_normalized.shape[0]
y_hat = np.zeros(n)

for i in range(n):
    y_hat[i] = np.dot(w_norm, X_normalized[i]) + b_norm

# plot predictions versus target values over original features
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range((len(ax))):
    ax[i].scatter(X_train[:, i], y_train, label='target')
    ax[i].scatter(X_train[:, i], y_hat, label='predict')
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel('price')
ax[0].legend()

fig.suptitle('target versus prediction using z-score normalized model')
plt.show()
```

### Prediction

Predict the price of a house with 1200 sqft, 3 bedrooms, 1 floor, 40 years old

```python
x_house = np.array([1200, 3, 1, 40])
x_house_norm = (x_house - mu) / sigma
print(x_house_norm)
predicted_y = np.dot(w_norm, x_house_norm) + b_norm
print('predicted price: $', predicted_y * 1000)
```
