# Feature Scaling and Learning Rate

```python
import numpy as np 
import matplotlib.pyplot as plt
from utils import load_house_data, gradient_descent_houses
```

## Housing Price Prediction

Given a dataset containing many training examples with four features

| Size (sqft) | Number of Bedrooms  | Number of floors | Age of  Home | Price (1000s dollars)  |   
| ------------| ------------------- |----------------- |--------------|----------------------- |  
| 952         | 2                   | 1                | 65           | 271.5                  |  
| 1244        | 3                   | 2                | 64           | 232                    |  
| 1947        | 3                   | 2                | 17           | 509.8                  |  
| ...         | ...                 | ...              | ...          | ...                    |

Predict price for other houses

### Dataset

```python
# load the dataset
X_train, y_train = load_house_data()
X_features = ['size(sqft)', 'bedrooms', 'floors', 'age']
```

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:, i], y_train)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("Price (1000s dollars)")
plt.show()
```

### Learning rate

```python
n, d = X_train.shape

initial_w = np.zeros((d,))
initial_b = 0.0

alpha = 9.9e-7
iterations = 10

w_out, b_out, J_hist_out = gradient_descent_houses(X_train, y_train, initial_w, initial_b, alpha, iterations)
print(f"w,b found by gradient descent: w: {w_out}, b: {b_out:0.2f}")
```

```python
# Plot cost versus iteration
plt.plot(J_hist_out)

plt.title('Cost vs. iteration')
plt.xlabel('iteration step')     
plt.ylabel('cost')   
```
