```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Housing Price Prediction

Given a dataset containing two training examples with only one feature

| Size (1000 sqft)  | Price (1000s of dollars) |
| ------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

Predict price for other houses

```python
# x_train: input variables (size in 1000 square feet)
# y_train: target values (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print('x_train =', x_train)
print('y_train =', y_train)
```

### Number of training examples

```python
# n: number of training examples
print('x_train.shape =', x_train.shape)

n = x_train.shape[0]
print('Number of training examples is', n)
```

```python
n = len(x_train)
print('Number of training examples is', n)
```

### $i^{th}$ training example

```python
# (x^i, y^i): ith training example
i = 0
print(f"(x^({i}), y^({i})) = ({x_train[i]}, {y_train[i]})")
```

### Visualize the data

```python
# Plot the data
plt.scatter(x_train, y_train, marker='x', c='r')

# Set the title
plt.title('House Prices')

# Set the axes label
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price (1000s of dollars)')

plt.show()
```

### Model function

$$ f_{w, b}(x) = wx + b $$

```python
def compute_model_output(x, w, b):
    
    """ 
    Computes the prediction of a linear model
    Args:
        x (ndarray(n,)) : input variables, n training examples
        w, b (scalar)   : model parameters
    Returns:
        f (ndarray(n,)) : model outputs
    """
    n = x.shape[0]
    f = np.zeros(n)
    for i in range(n):
        f[i] = w * x[i] + b

    return f
```

### Plot the prediction

```python
w = 200
b = 100

f = compute_model_output(x_train, w, b)

# Plot the model prediction
plt.plot(x_train, f, c='b', label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('House Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price (1000s of dollars)')

plt.legend()
plt.show()
```

### Prediction

Predict the price of a house with 1200 sqft

```python
def predict(x, w, b):
    """ 
	Predicts target value y of the input variable x
    Args:
		x (scalar)		: an input variable
        w, b (scalar)	: model parameters
    Returns:
		f (scalar)		: predicted y
    """
    return w * x + b
```

```python
x = 1.2
predicted_y = predict(x, w, b)
print(f'Price of a 1200 square feet house is ${predicted_y:.0f},000')
```
