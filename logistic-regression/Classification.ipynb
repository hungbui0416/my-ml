# Classification

```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

### Dataset

```python
x_train = np.array([0, 1, 2, 3, 4, 5])
y_train = np.array([0, 0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1 ,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
def plot_data(X, y, ax, pos_label='y=1', neg_label='y=0', s=80, loc='best'):
    pos = (y == 1)
    neg = (y == 0)

    ax.scatter(X[pos, 0], X[pos, 1], marker='x', s=s, c='r', label=pos_label)
    ax.scatter(X[neg, 0], X[neg, 0], marker='o', s=s, label=neg_label, facecolors='none',
               edgecolors='#0096ff', lw=3)
    ax.legend(loc=loc)
```

```python
pos = (y_train == 1)
neg = (y_train == 0)

fig, ax = plt.subplots(1, 2, figsize=(12, 3))

# single feature
ax[0].scatter(x_train[pos], y_train[pos], marker='x', s=80, c='r', label='y=1')
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label='y=0', facecolors='none',
              edgecolors='#0096ff', lw=3)
ax[0].set_ylim(-0.08, 1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one feature plot')
ax[0].legend()

# two features
plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_xlabel('$x_0$', fontsize=12)
ax[1].set_ylabel('$x_1$', fontsize=12)
ax[1].set_title('two features plot')

plt.show()
```
